```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import math
from google.colab.patches import cv2_imshow
from PIL import Image as im
img = cv.imread('/content/drive/MyDrive/PCVK/imanges/frozen.jpeg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
def convolution2d(image, kernel, stride, padding):

  image = np.pad(image, padding, mode='constant', constant_values=0)

  kernel_height, kernel_width = kernel.shape
  padded_height, padded_width = image.shape

  output_height = (padded_height - kernel_height) // stride + 1
  output_width = (padded_width - kernel_width) // stride + 1

  new_image = np.zeros((output_height, output_width)).astype(np.float32)

  # convolution process
  for y in range(0, output_height):
    for x in range(0, output_width):
      new_image[y][x] = np.sum(image[y * stride:y * stride + kernel_height, x * stride:x * stride + kernel_width] * kernel).astype(np.float32)
  return new_image
```

```python
img = cv.imread('/content/drive/MyDrive/PCVK/imanges/frozen.jpeg')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
# Image Sharpen Mean (rata-rata / AVG)
# Mean filter merupakan salah satu jenis low-pass filter 
# cara kerjanya dengan membuat nilai suatu piksel sesuai dengan matrix 3x3 jadi, perhitungan 1 piksel dari matrix tersebut yaitu 1/9.
# 1/9+1/9+1/9+1/9+1/9+1/9+1/9+1/9+1/9 = 9/9 = 1
kernel_sharpen = np.array([[1/9,1/9,1/9],
                           [1/9,1/9,1/9],
                           [1/9,1/9,1/9]])
```

```python
cv2_imshow(convolution2d(img_gray,kernel_sharpen,1,2))
```

```python
# Image Sharpen Median (nilai tengah)
# Median filter merupakan salah satu jenis low-pass filter 
# cara kerjanya dengan mengganti nilai suatu piksel pada citra asal dengan nilai median dari piksel tersebut dan lingkungan tetangganya.
kernel_sharpen2 = np.array([[1/9,1/9,1/9],
                           [1/9,1/18,1/9],
                           [1/9,1/9,1/9]])
```

```python
cv2_imshow(convolution2d(img_gray,kernel_sharpen2,1,2))
```

```python
# Image Sharpen Modus 
kernel_sharpen3 = np.array([[1/16,1/8,1/16],
                           [1/8,1/4,1/8],
                           [1/16,1/8,1/16]])
```

```python
# Hasil
print('Original')
cv2_imshow(img_gray)
print('Modus')
cv2_imshow(convolution2d(img_gray,kernel_sharpen3,1,2))
```

```python
# High Filtering akan menyebabkan tepi objek tampak lebih tajam dibandingkan sekitarnya.
# kernal_high = np.array([[1/9,1/2,1/9],
#                        [1/9,1/2,1/9],
#                        [1/9,1/2,1/9]])

high_filter = np.array([[-1, -1, -1],
                        [-1,  8, -1],
                        [-1, -1, -1]])
```

```python
# Hasil
print('Original')
cv2_imshow(img_gray)
print('High')
# cv2_imshow(convolution2d(img_gray,kernal_high,1,2))
cv2_imshow(convolution2d(img_gray,high_filter,1,2))
```

```python
# Low Pass Filtering sama dengan median
# kernal_low = np.array([[1/9,1/9,1/9],
#                        [1/9,1/2,1/9],
#                        [1/9,1/9,1/9]])

low_filter = np.array([[1, 1, 1],
                       [1, 1, 1],
                       [1, 1, 1]]) / 9
```

```python
# Hasil
print('Original')
cv2_imshow(img_gray)
print('Low')
# cv2_imshow(convolution2d(img_gray,kernal_low,1,2))
cv2_imshow(convolution2d(img_gray,low_filter,1,2))
```
