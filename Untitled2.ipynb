3. Membuat fungsi konvolusi untuk filtering Median

```python
def median_filtering(data, kernel):
  temp = []
  indexer = kernel //2
  data_hasil = []
  data_hasil = np.zeros((len (data), len (data[0])))
  for i in range(len(data)):
    for j in range(len(data[0])): 
      for z in range(kernel):
        if i + z - indexer < 0 or i + z - indexer > len(data) - 1:
          for c in range(kernel):
            temp.append(0)
        else:
          if j + z - indexer < 0 or j + indexer > len(data) - 1:
            temp.append(0)
          else:
            for k in range(kernel):
              temp.append(data[i + z - indexer][j + k - indexer])

      temp.sort()
      data_hasil[i][j] = temp[len(temp) // 2]
      temp = []
  return data_hasil
```

```python
cv2_imshow(median_filtering(img_gray, 3))
```
